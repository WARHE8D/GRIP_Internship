```python
import pandas as pd
import numpy as np
```

```python
match=pd.read_csv('C:/Users/SALMAN/Desktop/GRIP/matches.csv')
match.head()
```

```python
m_hist=pd.read_csv('C:/Users/SALMAN/Desktop/GRIP/deliveries.csv')
m_hist.head()
```

```python
match=match.drop(['season','umpire1','umpire2','umpire3'],axis=1)
match
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
city_le=le.fit_transform(match.city.astype(str))
```

```python
team1_le=le.fit_transform(match.team1.astype(str))
team2_le=le.fit_transform(match.team2.astype(str))
```

```python
toss_won_le=le.fit_transform(match.toss_winner.astype(str))
toss_call_le=le.fit_transform(match.toss_decision.astype(str))
```

```python
#target
win_le=le.fit_transform(match.winner.astype(str))
```

```python
#man of the match
MOTM_le=le.fit_transform(match.player_of_match.astype(str))
venu_le=le.fit_transform(match.venue.astype(str))
```

```python
match_le=pd.DataFrame(data=match)
```

```python
match_le['city_le']=city_le
match_le['team1_le']=team1_le
match_le['team2_le']=team2_le
match_le['toss_won_le']=toss_won_le
match_le['toss_call_le']=toss_call_le
match_le['winner_le']=win_le
match_le['MOMT_le']=MOTM_le
match_le['venue_le']=venu_le
```

```python
match_lEnc=match_le.drop(['city','team1','team2','toss_winner','toss_decision','result','dl_applied','winner','player_of_match','venue'],axis=1)
```

```python
match_lEnc.head()
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(match_lEnc.corr(),cmap='binary',annot=True)
plt.show()
```

```python
# as we cam see team 2 has higher chance of winning than team 1
# also the team who won the toss had greater chance of winning
```

```python
x=match_lEnc.iloc[:,2:]
x=x.drop(['winner_le'],axis=1)
x
```

```python
y=match_lEnc['winner_le']
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
xtrain.shape
```

```python
classifier=DecisionTreeClassifier()
```

```python
classifier.fit(xtrain,ytrain)
```

```python
ypred=classifier.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(ypred,ytest)
acc
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier2=RandomForestClassifier()
classifier2.fit(xtrain,ytrain)
```

```python
ypred2=classifier2.predict(xtest)
```

```python
acc2=accuracy_score(ypred2,ytest)
acc2
```

```python
# My model was 78% accurate
```
